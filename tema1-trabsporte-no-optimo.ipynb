```julia
import Pkg
Pkg.add("JuMP")
Pkg.add("CSV")
Pkg.add("Clp")
Pkg.add("GLPK")
Pkg.add("DataFrames")
Pkg.add("LinearAlgebra")

using JuMP, CSV, Clp, GLPK, DataFrames, LinearAlgebra

m = Model(with_optimizer(Clp.Optimizer))
#Variables: xij= por golon de combustible a ser distribuidas desde el proveedor i= 1, 2, 3  
# hasta el aeropuerto j= 1, 2, 3, 4
@variables(m, begin #Definimos las variables
        x11>=0 
        x12>=0             
        x13>=0             
        x21>=0             
        x22>=0             
        x23>=0             
        x31>=0             
        x32>=0             
        x33>=0             
         
end)  
# restricciones de Oferta
@constraint(m, x11+x12+x13<=320)        
@constraint(m, x21+x22+x23<=270)           
@constraint(m, x31+x32+x33<=190) 
 
# restricciones de Demanda
@constraint(m, x11+x21+x31==100)
@constraint(m, x12+x22+x32==180)
@constraint(m, x13+x23+x33==350)


@objective(m, Min, 92*x11+89*x12+90*x13+91*x21+91*x22+95*x23+87*x31+90*x32+92*x33)
print(m)
JuMP.optimize!(m)

obj_value=JuMP.objective_value(m)
x11=JuMP.value(x11)
x12=JuMP.value(x12)
x13=JuMP.value(x13)
x21=JuMP.value(x21)
x22=JuMP.value(x22)
x23=JuMP.value(x23)
x31=JuMP.value(x31)
x32=JuMP.value(x32)
x33=JuMP.value(x33)

println("Valor func objetivo_min=", obj_value)
println("x11 =", x11)
println("x12    =", x12)
println("x13    =", x13)


println("x21 =", x21)
println("x22    =", x22)
println("x23    =", x23)


println("x31 =", x31)
println("x32    =", x32)
println("x33    =", x33)
```
