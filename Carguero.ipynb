```julia
import Pkg
tempdir = mktempdir()
Pkg.activate(tempdir)
Pkg.add(["JuMP", "CSV","Clp","GLPK","DataFrames","LinearAlgebra"])

using JuMP, CSV, Clp, GLPK, DataFrames, LinearAlgebra

m = Model(with_optimizer(Clp.Optimizer))
#Variables: xj= toneladas
@variables(m, begin #Definimos las variables
        x1>=0
        x2>=0
        x3>=0
        x4>=0
        x5>=0
        x6>=0
        x7>=0
        x8>=0
        x9>=0
        x10>=0
end)  
# sa
@constraint(m, 100*x1+155*x2+50*x3+112*x4+70*x5+80*x6+60*x7+118*x8+110*x9+55*x10==700)        
@constraint(m, x1<=1)           
@constraint(m, x2<=1)
@constraint(m, x3<=1)
@constraint(m, x4<=1)
@constraint(m, x5<=1)
@constraint(m, x6<=1)
@constraint(m, x7<=1)
@constraint(m, x8<=1)
@constraint(m, x9<=1)
@constraint(m, x10<=1)

@objective(m, Max, 100*x1+155*x2+50*x3+112*x4+70*x5+80*x6+60*x7+118*x8+110*x9+55*x10)
print(m)
JuMP.optimize!(m)

obj_value=JuMP.objective_value(m)
x1=JuMP.value(x1)
x2=JuMP.value(x2)
x3=JuMP.value(x3)
x4=JuMP.value(x4)
x5=JuMP.value(x5)
x6=JuMP.value(x6)
x7=JuMP.value(x7)
x8=JuMP.value(x8)
x9=JuMP.value(x9)
x10=JuMP.value(x10)

println("Valor func objetivo_max=", obj_value)
```
